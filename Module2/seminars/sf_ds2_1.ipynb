````python
def int1_to_str(number: int) -> str:
    return chr(48 + number)

def int_to_str(number: int) -> str:
    a_ = abs(number)
    
    result = ''
    
    while a_ > 0:
        num = a_ % 10
        result += int1_to_str(num)
        a_ //= 10  # a_ = a_ // 10

    result = result[::-1]  # result = '-' * (number < 0) + result[::-1]

    if number < 0:
        result = '-' + result

    return result
````

````python
def int_to_str(number: int) -> str:
    if number < 10:
        return chr(48 + number)
    else:
        reminder = number % 10
        return int_to_str(number // 10) + chr(48 + reminder)
````

```python
import pandas as pd
import matplotlib.pylab as plt

df = pd.read_csv("./data/Electric_Car.csv")
```

```python
grouped_df = (
    df.groupby("Brand").size().sort_values().reset_index().rename(columns={0: "Count"})
)
grouped_df
```

```python
fig, ax = plt.subplots()
ax.grid()
ax.set_title("Количество моделей определенного бренда")
plt.barh(grouped_df["Brand"], grouped_df["Count"])
plt.show()
```
